```python
import numpy as np

def mysum(a):
    su = 0
    for i in a:
        su+=i
    return su
        

def mylength(a):
    le = 0
    for i in a:
        le+=1
    return le

def DET(x):
    s1  = x[1][1]*x[2][2] - x[2][1]*x[1][2] 
    s2 =  x[1][0]*x[2][2] - x[2][0]*x[1][2]
    s3 =  x[1][0]*x[2][1] - x[2][0]*x[1][1]
    res = x[0][0] * s1 - x[0][1] * s2 + x[0][2] * s3
    return res


x = np.array([float(x) for x in input("enter x values : ").split()])
y = np.array([ float(y) for y in input(" enter y values ").split()])
x2 = x*x
x3 = x2*x
x4 = x3*x
xy = x*y
x2y = x2*y

sx2 = mysum(x2)
sx3 = mysum(x3)
sx4 = mysum(x4)
sx2y = mysum(x2y)
sxy = mysum(xy)
sx = mysum(x)
sy = mysum(y)

n = mylength(x)

delta = DET([[n,sx,sx2],
            [sx,sx2,sx3],
            [sx2,sx3,sx4]])

delta1 = DET([[sy,sx,sx2],
            [sxy,sx2,sx3],
            [sx2y,sx3,sx4]])

delta2 = DET([[n,sy,sx2],
            [sx,sxy,sx3],
            [sx2,sx2y,sx4]])

delta3 = DET([[n,sx,sy],
            [sx,sx2,sxy],
            [sx2,sx3,sx2y]])


a  = delta1 / delta
b  = delta2 / delta
c  = delta3 / delta 

print(" the calculated parabola  y=",a,"+",b,"x+",c,"x^2 is the best fit . ")
```
